### ДЗ №1
Нужно посчитаь:

$$\int_0^2\frac{2+ cos(1+ \sqrt x)e^{0.5x}}{\sqrt{1+0.5x}}dx$$

Будем делать это методом Гаусса с полиномом Лежандра

#### Начинаем!

Примочки:

```python
import numpy as np
from scipy.special import roots_legendre

a = 0
b = 2
```

Главная героиня:

```python
def ff(x):
    
    f = (2 + np.cos(1 + np.sqrt (x)))/np.sqrt(1 + 0.5 * (x)) * np.exp(0.5 * (x))
    return f
```

Поменяем пределы на [-1; 1], чтобы машина смогла посчитать веса и точки для полинома:

```python
def lim_true(f, a, b, x):
    
    return (b - a)/2 * f((a + b)/2 + (a - b)*x/2)
```

Магия:

```python
def gauss(f, n):
    
    roots, weights = roots_legendre(n)
    res = np.sum(lim_true(ff, a, b, roots) * weights)
    return res
```

Вот, что получилось у нас с компьютером:

```python
print('Полином степени 5:\n', gauss(ff, 5))
print('\nПолином степени 10:\n', gauss(ff, 10))
print('\nПолином степени 50:\n', gauss(ff, 50))
```
